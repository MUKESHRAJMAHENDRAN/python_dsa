### 1. [Ransom Note](https://leetcode.com/problems/ransom-note/) - Easy  
**Problem Statement:**
Given two strings `ransomNote` and `magazine`, return `true` if `ransomNote` can be constructed from `magazine` and `false` otherwise.  
Each letter in `ransomNote` can only be used once from the magazine.

**Sample Input 1:**
```
ransomNote = "a", magazine = "b"
```

**Sample Output 1:**
```
false
```

**Sample Input 2:**
```
ransomNote = "aa", magazine = "ab"
```

**Sample Output 2:**
```
false
```

**Sample Input 3:**
```
ransomNote = "aa", magazine = "aab"
```

**Sample Output 3:**
```
true
```

````python
from collections import Counter

def canConstruct(ransomNote: str, magazine: str) -> bool:
    # Count the frequency of each character in ransomNote
    ransom_counter = Counter(ransomNote)
    # Count the frequency of each character in magazine
    magazine_counter = Counter(magazine)

    # Check if each character in ransomNote can be constructed from magazine
    for char, count in ransom_counter.items():
        # If magazine does not have enough of 'char', return False
        if magazine_counter[char] < count:
            return False

    # If we have enough of every required character, return True
    return True

# Sample Test Case
print(canConstruct("a", "b"))  # Output: False
print(canConstruct("aa", "ab"))  # Output: False
print(canConstruct("aa", "aab"))  # Output: True
````

### 2. [Insert Delete GetRandom O(1)](https://leetcode.com/problems/insert-delete-getrandom-o1/) - Medium  
**Problem Statement:**
Design a data structure that supports the following operations:

- `insert(val)`: Inserts an item `val` into the collection. Returns `true` if the item was successfully inserted, `false` if the item was already present.
- `remove(val)`: Removes an item `val` from the collection. Returns `true` if the item was successfully removed, `false` if the item was not found.
- `getRandom()`: Returns a random element from the collection. Each element must have the same probability of being returned.

**Sample Input 1:**
```
obj = RandomizedCollection()
obj.insert(1)
obj.insert(1)
obj.remove(1)
obj.getRandom()
```

**Sample Output 1:**
```
true, true, false, 1
```

````python
import random

class RandomizedSet:
    def __init__(self):
        # A list to store elements
        self.data = []
        # A dictionary to store the index of each element
        self.val_to_index = {}

    def insert(self, val: int) -> bool:
        # If val is already in set, return False
        if val in self.val_to_index:
            return False
        # Otherwise, add it to the list and the dictionary
        self.data.append(val)
        self.val_to_index[val] = len(self.data) - 1
        return True

    def remove(self, val: int) -> bool:
        # If val is not present, return False
        if val not in self.val_to_index:
            return False
        # Get index of the element to be removed
        index = self.val_to_index[val]
        # Move the last element to the place of the element to remove
        last_element = self.data[-1]
        self.data[index] = last_element
        self.val_to_index[last_element] = index
        # Remove the last element from list
        self.data.pop()
        # Remove the element from dictionary
        del self.val_to_index[val]
        return True

    def getRandom(self) -> int:
        # Return a random element from the list
        return random.choice(self.data)

# Sample Test Case
randomSet = RandomizedSet()
print(randomSet.insert(1))  # Output: True
print(randomSet.remove(2))  # Output: False
print(randomSet.insert(2))  # Output: True
print(randomSet.getRandom())  # Output: Randomly 1 or 2
print(randomSet.remove(1))  # Output: True
print(randomSet.insert(2))  # Output: False
print(randomSet.getRandom())  # Output: 2
````

### 3. [First Missing Positive](https://leetcode.com/problems/first-missing-positive/) - Hard  
**Problem Statement:**
Given an unsorted integer array `nums`, find the smallest missing positive integer.

You must implement an algorithm that runs in O(n) time and uses constant extra space.

**Sample Input 1:**
```
nums = [1, 2, 0]
```

**Sample Output 1:**
```
3
```

**Sample Input 2:**
```
nums = [3, 4, -1, 1]
```

**Sample Output 2:**
```
2
```

**Sample Input 3:**
```
nums = [7, 8, 9, 11, 12]
```

**Sample Output 3:**
```
1
```

````python
def firstMissingPositive(nums) -> int:
    n = len(nums)

    # Base case
    if 1 not in nums:
        return 1

    # If length of nums is 1, and it contains only 1
    if n == 1:
        return 2

    # Replace negative numbers, zeros, and numbers greater than n with 1s
    for i in range(n):
        if nums[i] <= 0 or nums[i] > n:
            nums[i] = 1

    # Use the index as a way to mark presence of numbers
    for i in range(n):
        a = abs(nums[i])
        # Use negative sign as a marker
        if a == n:
            nums[0] = -abs(nums[0])
        else:
            nums[a] = -abs(nums[a])

    # First positive number's index is the answer
    for i in range(1, n):
        if nums[i] > 0:
            return i

    # If nums[0] is positive, then n was not present
    if nums[0] > 0:
        return n

    return n + 1

# Sample Test Case
print(firstMissingPositive([1, 2, 0]))  # Output: 3
print(firstMissingPositive([3, 4, -1, 1]))  # Output: 2
print(firstMissingPositive([7, 8, 9, 11, 12]))  # Output: 1
````
